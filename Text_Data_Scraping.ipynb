```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
session = requests.Session()
session.headers['User-Agent']
```

```python
my_headers = {"User-Agent": "Mozilla/5.0 (X11; CrOS x86_64 14685.0.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.4992.0 Safari/537.36",
              "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange"}
```

```python
input_file = pd.read_excel('/content/drive/MyDrive/Text Analysis/Input.xlsx')
input_file.head()
```

```python
contents = []

for url in input_file['URL']:
  data = session.get(url, headers=my_headers)
  soup = BeautifulSoup(data.content, 'html.parser')

  body = ''
  for text in soup.find_all('p'):
    body = f'{body} {str(text.text)}'

  contents.append(body)
```

```python
df = pd.merge(input_file, pd.DataFrame(contents), left_index = True, right_index = True, how = "left")
df.columns = ['URL_ID', 'URL', 'Text_Content']
df.head()
```

```python
df.to_csv('/content/drive/MyDrive/Text Analysis/Content_file.csv', index=False)
```
